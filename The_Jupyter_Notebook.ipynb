```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Project A: Wealth Accumulation & Compounding Analysis
# Parameters based on your specific investment strategy

def generate_investment_data(initial_capital=10000, monthly_gift=5000, annual_rate=0.075, years=5):
    # Calculations for daily compounding
    days = years * 365
    daily_rate = annual_rate / 365

    current_date = datetime(2025, 4, 1) # Starting from your planned start date
    balance = initial_capital
    records = []

    for day in range(days):
        # Monthly capital injection (simulating the monthly gift from your father)
        if current_date.day == 1 and day > 0:
            balance += monthly_gift

        # Daily compounding interest calculation
        interest = balance * daily_rate
        balance += interest

        records.append({
            "Date": current_date.strftime('%Y-%m-%d'),
            "Portfolio_Value": round(balance, 2),
            "Daily_Interest": round(interest, 2)
        })
        current_date += timedelta(days=1)

    return pd.DataFrame(records)

# Generate the dataset
df = generate_investment_data()

# Show the first few rows to verify
print(df.head())

# Save for Power BI
df.to_csv('investment_data.csv', index=False)
```
