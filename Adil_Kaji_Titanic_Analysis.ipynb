```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the data
df = sns.load_dataset('titanic')

# 2. Clean the data (Wrangling)
df['age'] = df['age'].fillna(df['age'].mean())
df.drop(columns=['deck', 'embark_town', 'alive'], inplace=True)

# 3. Analyze Survival by Class
survival_by_class = df.groupby('pclass')['survived'].mean() * 100

# 4. Show the Results
print("--- Titanic Survival Analysis ---")
print(survival_by_class)

# 5. Create the Masterpiece Chart
survival_by_class.plot(kind='bar', color=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Survival Rate by Class (%)')
plt.ylabel('Percentage')
plt.xlabel('Class (1=Rich, 3=Poor)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Visualize the 'Class vs. Survival' trend
plt.figure(figsize=(8, 5))
sns.barplot(x='pclass', y='survived', data=df, palette='viridis')

# 2. Add professional labels
plt.title('Survival Probability by Passenger Class', fontsize=14)
plt.xlabel('Ticket Class (1 = 1st, 2 = 2nd, 3 = 3rd)', fontsize=12)
plt.ylabel('Survival Rate (Percentage)', fontsize=12)

# 3. Save the chart for your GitHub
plt.savefig('survival_by_class.png')
plt.show()
```

```python
# 1. Create a "Hue" chart to show Survival by Class AND Sex
plt.figure(figsize=(10, 6))
sns.barplot(x='pclass', y='survived', hue='sex', data=df, palette='magma')

# 2. Add professional formatting
plt.title('Survival Rate by Class and Gender', fontsize=16)
plt.xlabel('Ticket Class', fontsize=12)
plt.ylabel('Survival Probability', fontsize=12)
plt.legend(title='Gender')

# 3. Show the final visualization
plt.show()
```
