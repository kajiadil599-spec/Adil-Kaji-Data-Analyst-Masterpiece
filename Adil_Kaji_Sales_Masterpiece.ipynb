```python
import pandas as pd
import numpy as np

# 1. Create a synthetic Sales Dataset
np.random.seed(42)
data_size = 500
data = {
    'Order_ID': range(1001, 1001 + data_size),
    'Product_Category': np.random.choice(['Electronics', 'Furniture', 'Office Supplies', 'Clothing'], data_size),
    'Sales_Amount': np.random.uniform(50, 1500, data_size).round(2),
    'Discount_Percent': np.random.uniform(0, 0.3, data_size).round(2),
    'Shipping_Cost': np.random.uniform(5, 50, data_size).round(2),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], data_size)
}

df_sales = pd.DataFrame(data)

# 2. Save it as a CSV for your portfolio
df_sales.to_csv('ecommerce_sales_data.csv', index=False)
print("Project 2 Dataset Generated: 'ecommerce_sales_data.csv'")
df_sales.head()
```

```python
# 1. Calculate the actual Revenue (after discount)
df_sales['Actual_Revenue'] = df_sales['Sales_Amount'] * (1 - df_sales['Discount_Percent'])

# 2. Calculate Profit (assuming a 40% cost of goods, then subtracting shipping)
df_sales['Cost_of_Goods'] = df_sales['Sales_Amount'] * 0.40
df_sales['Net_Profit'] = df_sales['Actual_Revenue'] - df_sales['Cost_of_Goods'] - df_sales['Shipping_Cost']

print("New Metrics Created: Actual_Revenue and Net_Profit")
df_sales[['Product_Category', 'Sales_Amount', 'Net_Profit']].head()
```

```python
# Group by Region to see total Profit
region_performance = df_sales.groupby('Region')['Net_Profit'].sum().sort_values(ascending=False)

print("Profitability by Region:")
print(region_performance)

# Visualize it
region_performance.plot(kind='pie', autopct='%1.1f%%', figsize=(7,7), colors=['#4CAF50', '#2196F3', '#FFC107', '#F44336'])
import matplotlib.pyplot as plt
plt.title('Contribution to Total Profit by Region')
plt.show()
```

```python
import pandas as pd
import numpy as np
from google.colab import files

# 1. Generate the Dataset
np.random.seed(42)
data_size = 500
data = {
    'Order_ID': range(1001, 1001 + data_size),
    'Product_Category': np.random.choice(['Electronics', 'Furniture', 'Office Supplies', 'Clothing'], data_size),
    'Sales_Amount': np.random.uniform(50, 1500, data_size).round(2),
    'Discount_Percent': np.random.uniform(0, 0.3, data_size).round(2),
    'Shipping_Cost': np.random.uniform(5, 50, data_size).round(2),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], data_size)
}

df_sales = pd.DataFrame(data)

# 2. Add the Profit Metrics (The "Masterpiece" Logic)
df_sales['Actual_Revenue'] = (df_sales['Sales_Amount'] * (1 - df_sales['Discount_Percent'])).round(2)
df_sales['Cost_of_Goods'] = (df_sales['Sales_Amount'] * 0.40).round(2)
df_sales['Net_Profit'] = (df_sales['Actual_Revenue'] - df_sales['Cost_of_Goods'] - df_sales['Shipping_Cost']).round(2)

# 3. Save to CSV
df_sales.to_csv('ecommerce_sales_data.csv', index=False)

# 4. DOWNLOAD to your computer
files.download('ecommerce_sales_data.csv')

print("Check your downloads folder!")
```
